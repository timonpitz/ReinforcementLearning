## 2.2 Agent: wird von uns trainiert, eine optimale Policy zu entwicklen

Die **Policy (Strategie)** beschreibt für jeden möglichen Zustand, wie sich der Agent in dem Zustand verhält

## 2.3 Gängige Probleme
### Exploration vs Exploitation 

Exploration: Je schlechter die aktuelle Policy ist, desto nützlicher ist es, von ihr abweichende Actions zu erforschen <br> <br>
Exploitation: Je näher die aktuelle Policy and der opimalen Policy ist, desto nützlicher ist es, der Policy zu folgen 

### Delayed Rewards

Erst wenn der Agent am Ziel ist, wird er belohnt (Sparse Reward Setting)
* wenn der Reward zu schwierig ist zu erlangen, kann nie herausgefunden werden, welche Actions gut sind
* es kann nicht herausgefunden werden, welche Actions auf dem Weg zum Award gut sind
